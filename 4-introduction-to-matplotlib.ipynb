**The % symbol** is used as a prefix to indicate a Jupyter Notebook magic command. When used with Matplotlib as %matplotlib, it serves the purpose of configuring how Matplotlib plots are displayed within the Jupyter Notebook environment.

- %matplotlib inline: This magic command tells Jupyter to render Matplotlib plots directly within the notebook, below the code cell that generates them. It's useful for interactive data analysis and visualization within the notebook.

- %matplotlib notebook: This command enables interactive plots that allow zooming, panning, and other interactive features within the notebook itself. It provides a more dynamic plotting experience.

These magic commands make it easier to work with Matplotlib in Jupyter Notebook, enhancing the integration of data visualization into your data analysis workflow.

<br>

**%matplotlib inline**: 
- This Jupyter Notebook magic command instructs the notebook to display Matplotlib plots directly within the notebook interface, making it convenient for interactive data visualization.
- If we do not use %matplotlib inline in a Jupyter Notebook, Matplotlib will typically display plots in separate windows or pop-up dialog boxes, depending on your operating system and configuration
- Using %matplotlib inline ensures that Matplotlib plots are displayed directly within the notebook itself


**import matplotlib.pyplot as plt**: 
- This line imports the Matplotlib library's pyplot module and gives it the alias plt. This alias simplifies the process of creating and customizing plots using Matplotlib in your Python code.

# 4. Matplotlib Workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))  # (width, height)

# 3. Plot Data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & show (save the whole figure)
fig.savefig("exports/matplotlib/1-sample-plot.png")
```

# 5. Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

(There are many more plots)

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

### 5.1 Line Chart 

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

### 5.2 Bar Chart 

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew  butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())  # (x, y)
ax.set(title="Butter Store",
      ylabel="Price ($)");

# The semicolon should be applied to the last line where 'ax' is used
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));  # horizontal bar chart
```

### 5.3 Histogram

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### 5.4 Two options for subplots

#### Option 1

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

#### Option 2

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```
